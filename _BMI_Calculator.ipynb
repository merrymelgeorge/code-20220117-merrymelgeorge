```python
#Importing the required packages
import pandas as pd
import json
```

```python
#Loading the json file 
json_sample_data = json.load(open('testfile.json'))

#Creating a pandas dataframe from the json data
data = pd.DataFrame(json_sample_data)
data
```

```python
#Lambda function to calculate BMI
BMI = lambda wt,ht: wt/pow(ht/100,2) 

#Lambda function to find the BMI category based on the BMI value that is calculated dynamically
BMICat = lambda x,y: 'Underweight' if BMI(x,y)<=18.4 else ('Normal weight' if 18.5<=BMI(x,y)<=24.9 else ('Overweight' if 25<=BMI(x,y)<=29.9 else ('Moderately obese' if 30<=BMI(x,y)<=34.9 else ('Severly Obese' if 35<=BMI(x,y)<=39.9 else 'Very severly obese'))))

#Lambda function to find the Health Risk category based on the BMI value that is calculated dynamically
HRisk = lambda x,y: 'Malnutrition risk' if BMI(x,y)<=18.4 else ('Low risk' if 18.5<=BMI(x,y)<=24.9 else ('Enhanced risk' if 25<=BMI(x,y)<=29.9 else ('Medium risk' if 30<=BMI(x,y)<=34.9 else ('High risk' if 35<=BMI(x,y)<=39.9 else 'Very high risk'))))

#Adding the 'BMI','BMI_Category','Health Risk' columns to the dataframe
data[['BMI','BMI_Category','Health Risk']] = [[BMI(x,y),BMICat(x,y),HRisk(x,y)] for x,y in data[['WeightKg','HeightCm']].itertuples(index=False)]
```
